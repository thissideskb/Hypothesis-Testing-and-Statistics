# Question 1
What is hypothesis testing in statistics?

**Answer:**  
Hypothesis testing is a method to decide whether the evidence from sample data supports a certain claim about a population.  
- We start with an assumption (null hypothesis), collect data, and then check if the data provides enough evidence to reject that assumption.  
**Example:** Testing if the average height of students is really 170 cm or not.

# Question 2
What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Answer:**  
- **Null Hypothesis (H₀):** The assumption that there is no effect or difference. Example: H₀: Mean = 50.  
- **Alternative Hypothesis (H₁):** The statement that contradicts H₀, meaning there is some effect or difference. Example: H₁: Mean ≠ 50.  
Difference: H₀ assumes “no change,” while H₁ assumes “change exists.”

# Question 3
Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

**Answer:**  
The **significance level (α)** is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01.  
- If the p-value ≤ α, we reject H₀.  
- If the p-value > α, we fail to reject H₀.  
**Role:** It sets the cutoff for how much evidence we require to declare results as “statistically significant.”

# Question 4
What are Type I and Type II errors? Give examples of each.

**Answer:**  
- **Type I Error (False Positive):** Rejecting H₀ when it is true.  
  Example: Concluding a medicine works when it actually doesn’t.  
- **Type II Error (False Negative):** Failing to reject H₀ when it is false.  
  Example: Concluding a medicine doesn’t work when it actually does.

# Question 5
What is the difference between a Z-test and a T-test? Explain when to use each.

**Answer:**  
- **Z-test:** Used when the sample size is large (n > 30) and the population variance is known.  
- **T-test:** Used when the sample size is small (n ≤ 30) and the population variance is unknown.  
Both compare sample statistics with population parameters, but the T-test accounts for extra uncertainty in small samples.



```python
# Question 6
# Generate a binomial distribution with n=10, p=0.5, and plot its histogram.

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 samples
data = np.random.binomial(n=10, p=0.5, size=1000)

print("Sample Mean:", np.mean(data))
print("Sample Variance:", np.var(data))

plt.hist(data, bins=11, edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()

# Explanation:
# We generated outcomes of 1000 experiments, each with 10 trials and p=0.5.
# The histogram shows how often each number of successes occurs.
```

```python
# Question 7
# Implement hypothesis testing using Z-statistics for the given sample dataset.

import numpy as np
from math import sqrt

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

xbar = np.mean(sample_data)       # sample mean
sigma = 1                         # assume population std = 1
n = len(sample_data)
mu0 = 50                          # null hypothesis mean

# Z-statistic
z = (xbar - mu0) / (sigma / sqrt(n))

print("Sample Mean:", round(xbar, 3))
print("Z-statistic:", round(z, 3))

# For 95% confidence, critical Z = ±1.96
if abs(z) > 1.96:
    print("Reject H0: Significant difference from 50")
else:
    print("Fail to reject H0: No significant difference from 50")

# Explanation:
# If Z lies outside [-1.96, +1.96], we reject H0 at 5% level.
```

```python
# Question 8
# Simulate normal data and calculate 95% CI for the mean.

import numpy as np
import matplotlib.pyplot as plt
import math

np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=200)

xbar = np.mean(data)
s = np.std(data, ddof=1)
n = len(data)

# 95% CI using normal approx
se = s / math.sqrt(n)
z_crit = 1.96
lower = xbar - z_crit * se
upper = xbar + z_crit * se

print("Mean:", round(xbar, 2))
print("95% CI: [", round(lower, 2), ",", round(upper, 2), "]")

plt.hist(data, bins=30, edgecolor='black')
plt.title("Normal Distribution Simulation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Explanation:
# The CI gives a range where the true population mean likely lies with 95% confidence.
```

```python
# Question 9
# Function to calculate Z-scores and visualize standardized data.

import numpy as np
import matplotlib.pyplot as plt

def compute_zscores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    zscores = [(x - mean) / std for x in data]
    return zscores

# Example dataset
data = [45, 50, 55, 60, 65, 70, 75, 80, 85]

zscores = compute_zscores(data)

print("Z-scores:", [round(z, 2) for z in zscores])

plt.hist(zscores, bins=5, edgecolor='black')
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

# Explanation:
# A Z-score tells how many standard deviations a value is from the mean.
# Example: Z=+1 means 1 std above the mean, Z=-1 means 1 std below.
```
